# Plotting Examples
Simple examples of how to plot data using matplotlib

```python
# import dependencies
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = [0,1,2]
y = [0,2,4]
```

```python
x2 = np.linspace(0, 2, 20)
y2 = x2 ** 2
```

```python
# create figure
fig = plt.figure()
axes = fig.add_subplot(111)

# plot the data
axes.plot(x, y)
axes.plot(x2, y2)

# show the plot
plt.show()
```

```python
# change color of plots
fig2 = plt.figure()
axes = fig2.add_subplot(111)

# set title
axes.set_title('Color Examples')

# add labels
axes.set_xlabel('x')
axes.set_ylabel('y')

# plot the data
axes.plot(x, y, color='green', linestyle='dotted', marker='o', markersize=6)
axes.plot(x2, y2, color='red', linestyle='dashed')

# show the plot
plt.show()
```
